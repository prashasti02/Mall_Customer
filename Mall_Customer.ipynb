# **Data Analysis**


## Histogram 

Histogram is a value distribution plot of numerical columns.It is used to summarize discrete or continuous data that are measured on an interval scale.

 We used histogram to show clearly what is the age of customers.


```python
plt.hist(dataset_url['Age'], bins=100)
plt.show()
```

## Bar plot


A bar plot a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.

We use this graph to see the distribution of number of customers in each age group. It is more clearly shown and well distributed.

```python
age18_29 = dataset_url.Age[(dataset_url.Age >= 18) & (dataset_url.Age <= 29)]
age30_39 = dataset_url.Age[(dataset_url.Age >= 30) & (dataset_url.Age <= 39)]
age40_49 = dataset_url.Age[(dataset_url.Age >= 40) & (dataset_url.Age <= 49)]
age50_59 = dataset_url.Age[(dataset_url.Age >= 50) & (dataset_url.Age <= 59)]
age60_70 = dataset_url.Age[(dataset_url.Age >= 60) & (dataset_url.Age <= 70)]

x = ["18-29","30-39","40-49","50-59","60-70"]
y = [len(age18_29.values),len(age30_39.values),len(age40_49.values),len(age50_59.values),len(age60_70.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=x, y=y, palette="rainbow")
plt.title("Number of Customer and Ages")
plt.xlabel("Age")
plt.ylabel("Number of Customer")
plt.show()
```

## Box plot

A boxplot is a standardized way of displaying the dataset based on a five-number summary: the minimum, the maximum, the sample median, and the first and third quartile.



```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
sns.boxplot(x=dataset_url['Gender'],y=dataset_url["Spending Score (1-100)"], color="violet")
plt.subplot(1,2,2)
sns.boxplot(x=dataset_url['Gender'],y=dataset_url["Annual Income (k$)"], color="lightgreen")
plt.show()
```

# Pie Chart


The pie chart is gives  information about the percentage presence of each category in data.

We can clearly see that the female population is more than male population.

```python
dataset_url['Gender'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

# Distplot

Distplot accepts the data variable as an argument and returns the plot with the density distribution.

```python
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')
sns.distplot(dataset_url['Annual Income (k$)'])
plt.title('Distribution of Annual Income (k$)', fontsize = 10)
plt.xlabel('Range of Annual Income (k$)')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(10, 6))
sns.set(style = 'whitegrid')
sns.distplot(dataset_url['Spending Score (1-100)'],color='red')
plt.title('Distribution of Spending Score', fontsize = 10)
plt.xlabel('Range of Spending Score (1-100)')
plt.ylabel('Count')
plt.show()
```

# Scatter Plot

A scatter plot is a chart type that is normally used to observe and visually display the relationship between variables.



```python
sns.scatterplot(dataset_url["Spending Score (1-100)"],
                dataset_url["Annual Income (k$)"], 
                hue=dataset_url["Gender"])
```
